Q3. Que penser de la colonne `duration` par rapport à la *target* `y` ? Est-il réaliste de considérer que nous aurons cette valeur pour la prédiction ? Qu'en conclure, comment doit-on traiter la colonne `duration` ?

##### **Réponse**

La colonne `duration` représente la durée de l'appel en secondes pour chaque client, et elle est liée à la campagne téléphonique. Cependant, lorsqu'on l'analyse par rapport à la variable cible `y` (souscription ou non à un dépôt à terme), il y a quelques points à considérer.

##### 1. **Analyse de la relation entre `duration` et `y`** :
La colonne `duration` est souvent fortement corrélée avec la variable cible, car plus un appel dure longtemps, plus il est probable que le client soit intéressé et finisse par souscrire. Cela peut donner une impression qu'il s'agit d'un excellent prédicteur, mais cela pose une question importante.

##### 2. **Problème pour la prédiction** :
La colonne `duration` reflète le résultat de l'interaction avec le client (l'appel). Si l'appel est court, cela peut signifier un désintérêt immédiat de la part du client, tandis qu'un appel long peut indiquer une conversion potentielle. Cependant, **cette donnée n'est disponible qu'après l'appel**. 

- **Problème** : Pour faire une prédiction **avant l'appel**, la durée ne peut pas être utilisée, car elle est obtenue au moment de la campagne elle-même. Cela introduirait un biais si on l'utilisait dans un modèle prédictif pour anticiper la souscription avant d'effectuer l'appel.

##### 3. **Conclusion** :
- Il n'est pas réaliste de considérer la variable `duration` comme une feature utilisable pour prédire la souscription à un dépôt à terme **avant l'appel**, car cette information n'est disponible qu'après que l'interaction ait eu lieu.
- **Traitement suggéré** : Il est préférable de **supprimer la colonne `duration`** lors de la phase d'entraînement et de prédiction, car elle introduirait un biais non réaliste pour la modélisation. Cette colonne reflète plus un résultat qu'un facteur prédictif.

<details>
<summary>Astuce</summary>

Vous pouvez implémenter les normalisations à la main, car elles sont assez simples à faire en NumPy. En revanche, pour les encodages, et en particulier one-hot, il peut être assez fastidieux de compter le nombre d'éléments uniques, puis de générer n colonnes et de mettre la bonne valeur pour chaque colonne. Vous pouvez utiliser le module [preprocessing de Scikit-learn](https://scikit-learn.org/stable/modules/preprocessing.html) pour vous faciliter la vie. Voir en particulier [OneHotEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html#sklearn.preprocessing.OneHotEncoder).

Scikit suppose que vous passerez en entrée votre jeu de données entier dans un encodeur (c'est-à-dire un tableau 2D : individus et *features*). Quand on veut convertir 1 seule colonne, il faut forcer la conversion en tableau 2D quand même (c'est le retour de *reshape* !). Si votre jeu de données a une shape `(n, k)` avec `n` lignes et `k` *features*, une unique colonne a une shape `(n,)`, on veut obtenir à la place `(n,1)`.

Voici un exemple pour vous aider :

```python
X = np.asarray([
    [1, 2, 3],    # 1er individu
    [4, 5, 6],    # 2e individu
])

print(X.shape)
> (2, 3)
# 2 individus, 3 features

# On veut la 2e colonne (d'indice 1)
column = X[:, 1]

print(column.shape)
> (2,)
# 2 individus (Tableau 1D !!!)

column = column.reshape(-1, 1)
print(column.shape)
> (2,1)
# Toujours 2 individus et toujours 1 seule feature mais cette fois en 2D !

encoder = OneHotEncoder(...)          # À vous de remplir cette partie ;-)
column_encoded = encoder....(column)  # Ici aussi !

print(column_encoded.shape)
> (2,2)
# Toujours 2 individus mais l'encodage prend 2 colonnes car OneHot : autant de colonnes que de valeurs uniques...
```

Notez que, une fois l'encodage effectué par `OneHotEncoder`, on obtient plusieurs colonnes : il faut donc les nommer ! On pourra utiliser la méthode [`OneHotEncoder.get_feature_names_out()`](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html#sklearn.preprocessing.OneHotEncoder.get_feature_names_out) pour générer les noms pour chaque colonne à partir du nom "originel" de la *feature*.

Par exemple, si la colonne s'appelait `"job"` et que l'on a les valeurs `["commercial", "prof", "étudiant"]`, alors `encoder.get_feature_names_out(["job"])` aura pour valeurs `["job_commercial", "job_prof", "job_étudiant"]`. Il vous reste alors à assigner chacune des colonnes encodées avec son nom correspondant dans votre DataFrame résultat.
</details>

````python
# Copie du DataFrame original pour manipulation
df_prepared = df_cleaned.copy()

### 1. One-Hot Encoding pour les variables catégorielles raisonnables
# Colonnes à encoder en One-Hot
onehot_columns = ['month', 'marital', 'education']

# Appliquer One-Hot encoding
df_prepared = pd.get_dummies(df_prepared, columns=onehot_columns, drop_first=True)

### 2. Encodage Logarithmique pour 'balance' et 'campaign' avec clipping
# Clipping des valeurs inférieures à 1 avant la transformation
df_prepared['balance_clipped'] = df_cleaned['balance'].clip(lower=1)

# Appliquer la transformation logarithmique sur les valeurs clippées
df_prepared['balance_log'] = np.log1p(df_prepared['balance_clipped'])

# Même traitement pour 'campaign' si nécessaire
df_prepared['campaign_clipped'] = df_cleaned['campaign'].clip(lower=1)
df_prepared['campaign_log'] = np.log1p(df_prepared['campaign_clipped'])

# On peut supprimer les colonnes originales après encodage logarithmique
df_prepared = df_prepared.drop(columns=['balance', 'campaign', 'balance_clipped', 'campaign_clipped'])

### 3. Z-normalisation pour 'age'
scaler = StandardScaler()
df_prepared['age'] = scaler.fit_transform(df_prepared[['age']])

### 4. Encodage binaire pour des colonnes binaires (yes/no)
binary_columns = ['default', 'housing', 'loan']
df_prepared[binary_columns] = df_prepared[binary_columns].replace({'yes': 1, 'no': 0})

### 5. Affichage des distributions avant/après normalisation et transformation
# Création de subplots pour visualiser les distributions avant et après les transformations
fig, axes = plt.subplots(4, 2, figsize=(16, 12))

# Tracer 'age' avant et après normalisation
sns.histplot(df_cleaned['age'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age - Avant normalisation')
sns.histplot(df_prepared['age'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Age - Après Z-normalisation')

# Tracer 'balance' avant et après transformation logarithmique
sns.histplot(df_cleaned['balance'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Balance - Avant transformation')
sns.histplot(df_prepared['balance_log'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Balance - Après transformation logarithmique')

# Tracer 'campaign' avant et après transformation logarithmique
sns.histplot(df_cleaned['campaign'], kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Campaign - Avant transformation')
sns.histplot(df_prepared['campaign_log'], kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Campaign - Après transformation logarithmique')

# Tracer 'day' avant et après One-Hot encoding (le One-Hot n'est pas représenté ici, car il génère des colonnes)
sns.histplot(df_cleaned['day'], kde=True, ax=axes[3, 0])
axes[3, 0].set_title('Day - Avant encodage')
sns.histplot(df_prepared['day'], kde=True, ax=axes[3, 1])
axes[3, 1].set_title('Day - Après encodage')

plt.tight_layout()
plt.show()

# Affichage des premières lignes du DataFrame préparé
print(df_prepared.head())
````
